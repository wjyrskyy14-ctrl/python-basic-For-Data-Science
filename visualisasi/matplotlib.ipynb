## MATPLOTLIB

Macam macam CHART

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1. LINE CHART

CUSTOMIZATION

```python
x = np.array([2020, 2021, 2022, 2023])
y1 = np.array([10, 15, 7, 12])
y2 = np.array([5, 8, 6, 9])
```

```python
line_style = dict(marker='o', # Menambahkan marker bulat pada setiap titik data
               markersize=8, # Ukuran marker
               markerfacecolor='red', # Warna isi marker
               markeredgecolor='red', # Warna tepi marker
               linestyle='solid', # Jenis garis
               linewidth=2) # Ketebalan garis
```

```python
plt.plot(x, y1, color='blue', **line_style);
plt.plot(x, y2, color='green', **line_style);

##LABELING

# Menambahkan label pada sumbu x dengan kustomisasi font
plt.xlabel("tahun", fontsize=12,
                    fontweight='bold',
                    color='black')

# Menambahkan label pada sumbu y dengan kustomisasi font
plt.ylabel("jumlah mahasiswa", fontsize=12,
                    fontweight='bold',
                    color='black');

plt.xticks(x); # Menentukan posisi ticks pada sumbu x (tahun nya jadi integer)
plt.grid(); # Menambahkan grid pada grafik
plt.show(); # Menampilkan grafik
```

## 2. BARCHART

```python
kategori  = ['buah', 'sayur', 'daging', 'serealia']
jumlah = [25, 30, 15, 10]
```

```python
plt.bar(kategori, jumlah, color= 'blue')
plt.title('jumlah konsumsi makanan') #menambahkan judul bar
plt.xlabel(' makanan')
plt.ylabel('jumlah konsumsi(kg)')

plt.show()
```

## 3. PIE CHART

```python
merek = ['Toyota', 'Daihatsu', 'Honda', 'Suzuki']
jumlah_penjulan = [50, 60, 15, 5]
colors = ['blue', 'orange', 'green', 'red']
```

```python
plt.pie(jumlah_penjulan, labels=merek, # Menambahkan label pada setiap irisan
                         autopct='%1.1f%%', # Menampilkan persentase pada setiap irisan
                         colors = colors, # Warna pada setiap irisan
                         explode = [0, 0.1, 0, 0]) # Memisahkan irisan kedua (Daihatsu)

plt.title('Penjualan Mobil Tahun 2023') # Menambahkan judul pada grafik pie chart
plt.show() # Menampilkan grafik pie chart
```

## 4. SCATTER CHART

```python
x1 = np.array([0,1,2,1,4,8,9,7]) #jam belajar
y1 = np.array([35, 50, 40, 37, 65, 80, 95, 69]) #nilai ujian

x2 = np.array([9, 7, 4, 6, 1, 2, 0, 3]) #jam belajar
y2 = np.array([95, 75, 50, 70, 35, 25, 15, 40]) #nilai ujian
```

```python
# Membuat scatter plot untuk kelas A
plt.scatter(x1, y1, label= 'kelas A', # Menambahkan label untuk kelas A
                    color='blue', # Warna biru untuk titik data kelas A
                    marker='o') # # Marker berbentuk lingkaran untuk kelas A

# Membuat scatter plot untuk kelas B
plt.scatter(x2, y2, label= 'kelas B', # Menambahkan label untuk kelas B
                    color='red',  # Warna merah untuk titik data kelas B
                    marker='o') # Marker berbentuk silang untuk kelas B

plt.title('nilai  ujian ')
plt.xlabel('jam belajar')
plt.ylabel('nilai ujian')
plt.legend() # Menampilkan legenda untuk membedakan kelas A dan B

plt.show()
```
